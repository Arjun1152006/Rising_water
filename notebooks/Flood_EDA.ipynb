# Rising Waters: Exploratory Data Analysis (EDA)
### A Machine Learning Approach to Flood Prediction

This notebook performs a comprehensive exploratory analysis on our flood prediction dataset. The dataset contains environmental, atmospheric, and rainfall parameters. Our target is to predict whether a flood is expected (`Flood Label` = 1) or not (`Flood Label` = 0).

## 1. Import Libraries and Load Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Load the dataset
df = pd.read_csv('../dataset/flood.csv')
df.head()
```

## 2. Dataset Information
We check the number of rows, columns, data types, and check for missing and duplicate values.

```python
# Basic info
print("=== Dataset Info ===")
df.info()

# Check for missing values
print("\n=== Missing Values ===")
missing = df.isnull().sum()
print(missing[missing > 0])

# Check for duplicates
print(f"\nDuplicate rows found: {df.duplicated().sum()}")
```

## 3. Statistical Summary
Review descriptive statistics to understand centers, ranges, and variations.

```python
df.describe()
```

## 4. Data Visualizations

### 4.1 Target Variable Count Plot
Checking the class distribution (balance) of the target variable `Flood Label`.

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Flood Label', data=df, palette='Blues')
plt.title('Distribution of Flood Labels')
plt.show()
```

### 4.2 Correlation Matrix Heatmap
Understand how features relate to one another and to the target label. High positive correlation indicates features that influence flood occurrences.

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()
```

### 4.3 Feature Distribution Plots
We display histograms and density estimates of key predictors, like `River Level` and `Seasonal Rainfall`.

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.histplot(df['River Level'], kde=True, ax=axes[0], color='skyblue')
axes[0].set_title('River Level Distribution')

sns.histplot(df['Seasonal Rainfall'], kde=True, ax=axes[1], color='salmon')
axes[1].set_title('Seasonal Rainfall Distribution')
plt.tight_layout()
plt.show()
```

### 4.4 Outlier Detection Boxplots
Boxplots help detect extreme outliers in key features.

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.boxplot(y='River Level', x='Flood Label', data=df, ax=axes[0], palette='Blues')
axes[0].set_title('River Level by Flood Target')

sns.boxplot(y='Seasonal Rainfall', x='Flood Label', data=df, ax=axes[1], palette='Oranges')
axes[1].set_title('Seasonal Rainfall by Flood Target')
plt.tight_layout()
plt.show()
```

### 4.5 Top Feature Correlations
Let's rank features by their absolute correlation with the target `Flood Label`.

```python
correlations = df.corr()['Flood Label'].sort_values(ascending=False)
print("Correlation with Flood Label:")
print(correlations)
```
